```python
## Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style()
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

## Scenerio
You work in a call center and the primary responsibility of your organization log the call and resolve the customer facing issue. Recently some employee from your organization have left and moved to some other company. As as result you feel there has been significant decline in the productivity of the organization, you decide to pitch hiring manager to hire more staff explaining the scenerio with some visual plots.

### 1. Stacked Bar Plot

```python
tickets = pd.DataFrame({'Month':['January','February','March','April','May','June','July','August','September','October','November','December'],
                   'Ticket_Volume_Received':[160,184,241,149,180,161,132,202,160,139,149,177],
                   'Ticket_Volume_Processed':[160,184,237,148,181,150,123,156,126,104,124,140]})
```

```python
tickets.head()
```

```python
fig = plt.figure(figsize=(15,8))

# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.barplot(x = tickets.index, y = tickets['Ticket_Volume_Received'], data = tickets , label = 'Calls Logged',
            color='#4848FF')
sns.barplot(x = tickets.index, y = tickets['Ticket_Volume_Processed'], data = tickets, label = 'Calls Resolved',
            color='#FF6800')
plt.title('Monthly Volume of IT calls Logged and Resolved', fontsize=14, ha = 'center')

# Put a legend below current axis
plt.legend(loc='upper center', bbox_to_anchor = (0.5, -0.08), fancybox = True, shadow = False, ncol=5, fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300)
plt.xticks(range(0, 12), tickets["Month"].tolist())
plt.show()
```

This plot could be one of the go to option but certainly not the best one. Let's transform this plot in a way that could be more understandable.

## Perfect Visualization
### Line Plot

```python
fig = plt.figure(figsize=(15,8))
months = tickets["Month"].tolist()
months_3 = [m[:3] for m in months]

ax = fig.add_subplot(111)

# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.lineplot(x = tickets.index, y = tickets['Ticket_Volume_Received'], data = tickets,
            marker ='o', markersize = 10, lw = 4, color='#4848FF', ax = ax)
sns.lineplot(x = tickets.index, y = tickets['Ticket_Volume_Processed'], data = tickets,
            marker ='o', markersize = 10, lw = 4, color='#FF6800', ax = ax)
plt.title('Monthly Volume of IT calls Logged and Resolved', ha = 'center', fontsize=22, color='#616161' )
sns.despine()

# Change the colors of the left and bottom borders (fade into the background)
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(12.6, 180, 'Logged', ha='right', va='top', fontsize=22, color='#4848FF')
ax.text(12.8, 145, 'Resolved', ha='right', va='top', fontsize=22, color='#FF6800')

plt.ylabel('Volume', fontsize=16, color='#616161')
plt.ylim(0,300)
plt.xticks(range(0, 12), months_3)
plt.show()
```

I am sure now it looks much better. Lets see what are the things we have done.
1. Take only 3 character of months in x axis.
2. Remove border
3. Fade the borders into the background
4. Add labels on line plots
5. Make the title bigger and gray.
6. Remove the grid lines

## Storytelling with data

```python
y = [202,160,139,149,177]
x = [7, 8, 9, 10, 11]
n = y.copy()

y1 = [156,126,104,124,140]
x1 = [7, 8, 9, 10, 11]
n1 = y1.copy()

markers_on = [202,160,139,149,177]

fig = plt.figure(figsize=(15,8))
months = tickets["Month"].tolist()
months_3 = [m[:3] for m in months]

ax = fig.add_subplot(111)

# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.lineplot(x = tickets.index, y = tickets['Ticket_Volume_Received'], data = tickets,
            markersize = 10, lw = 4, color = '#808080', ax = ax)
sns.lineplot(x = tickets.index, y = tickets['Ticket_Volume_Processed'], data = tickets,
            markersize = 10, lw = 4, color = '#000088', ax = ax)
sns.lineplot(x, y, marker = 'o', markersize = 10, color = '#808080', ax = ax)
sns.lineplot(x1, y1, marker='o', markersize=10, color = '#000088', ax = ax)

plt.suptitle('Please approve the hire of 2 employees', fontsize = 24, ha = 'right', va = 'top', color = '#000088', y = 1)
sns.despine()


# Change the colors of the left and bottom borders (fade into the background)
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(12.6, 180, 'Logged', ha='right', va='top', fontsize=20, color='#808080')
ax.text(12.8, 145, 'Resolved', ha='right', va='top', fontsize=20, color='#000088')

plt.annotate("2 employees quit in May. We nearly kept up with incoming calls\nin the following two months,\
 but fell behind with the increase in\nAug and haven't been able to catch up since.",
             xy=(4,60),xytext=(4,260),color='#616161',fontsize=18)

plt.ylabel('Number of Calls',fontsize=16,color='#616161',ha='left',va='bottom')
plt.ylim(0,300)
plt.axvline(x=4, ymin=0, ymax=0.85, color='#616161', alpha=0.3)
plt.xticks(range(0, 12), months_3)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]),fontsize=16)
    

for i, txt in enumerate(n1):
    ax.annotate(txt, (x1[i], y1[i]),fontsize=16)
    
plt.show()
```
